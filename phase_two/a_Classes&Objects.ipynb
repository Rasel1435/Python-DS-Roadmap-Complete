## Classes & Objects
Classes & Objects (The Blueprint of Coding)

**What is a Class & Object? (কি এটা?)**

আসসালামু আলাইকুম Everyone! Phase 1-এ আমরা ছোট ছোট কাজ করা শিখেছি। কিন্তু বাস্তব জীবনে আমরা যখন কোনো সফটওয়্যার বানাই, তখন আমাদের 'জিনিস' বা 'Object' নিয়ে চিন্তা করতে হয়।

একটি Class হলো একটি Blueprint বা নকশা। আর সেই নকশা দেখে যখন আমরা আসল কিছু তৈরি করি, সেটা হলো Object।

সহজ উদাহরণ: একটি বাড়ির "ইঞ্জিনিয়ারিং ম্যাপ" হলো Class। আর সেই ম্যাপ দেখে যখন পাড়ায় ১০টি আলাদা আলাদা বাড়ি বানানো হয়, তখন প্রতিটি বাড়ি হলো এক একটি Object।

**Why do we use it? (কেন এটা ব্যবহার করব?)**

1. **Real-world Modeling**: মানুষ, গাড়ি, ব্যাংক অ্যাকাউন্ট—এসব বাস্তব জিনিসকে কোডে রূপান্তর করা সহজ হয়।

2. **Data + Action Together**: একটি অবজেক্টের ভেতর তার তথ্য (Variables) এবং তার কাজ (Functions) একসাথে গুছিয়ে রাখা যায়।

3. **Scalability**: বড় প্রজেক্টে যখন হাজার হাজার ডাটা থাকে, তখন ক্লাস ছাড়া কোড সামলানো অসম্ভব।

**How does it work? (Self এবং init এর রহস্য)**
-   **__init__ (The Constructor)**: এটি হলো একটি স্পেশাল ফাংশন যা অবজেক্ট তৈরির সাথে সাথেই নিজে থেকে রান করে। এর কাজ হলো অবজেক্টের প্রাথমিক তথ্য (যেমন নাম, বয়স) সেট করা।

-   **self**: এটি হলো ওই নির্দিষ্ট অবজেক্টের নিজের পরিচয়। এটি পাইথনকে বলে দেয় যে আমরা "এই" নির্দিষ্ট অবজেক্টের কথা বলছি।

**Detailed Code Example**

```python
# ক্লাস তৈরি (Blueprint)
class Student:
    def __init__(self, name, roll):
        self.name = name   # তথ্য জমা রাখা
        self.roll = roll
    
    def display_info(self): # কাজ বা Method
        print(f"Student: {self.name}, Roll: {self.roll}")

# অবজেক্ট তৈরি (Instances)
s1 = Student("Sheikh", 101)
s2 = Student("Rasel", 102)

s1.display_info()
s2.display_info()
```

### Mini Project — "The Robot Factory"
আমরা একটি রোবট তৈরির ক্লাস বানাব যেখানে প্রতিটি রোবটের আলাদা নাম এবং কাজ থাকবে।

```python
# --- Video 12 Project: Robot Factory ---

class Robot:
    def __init__(self, name, color, power_level):
        self.name = name
        self.color = color
        self.power = power_level
    
    def introduce(self):
        print(f"Hello! I am {self.name}. My color is {self.color}.")
    
    def check_power(self):
        if self.power > 50:
            print(f"{self.name} is ready for battle! 💪")
        else:
            print(f"{self.name} needs charging... ⚡")

# রোবট বানানো
bot1 = Robot("RoboOne", "Red", 80)
bot2 = Robot("IronMan", "Gold", 30)

bot1.introduce()
bot1.check_power()

bot2.introduce()
bot2.check_power()
```
