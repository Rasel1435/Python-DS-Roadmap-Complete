## Inheritance (The Power of Legacy) 🧬

### What is Inheritance? (কি এটা?)

বাস্তব জীবনে যেমন একজন সন্তান তার বাবার থেকে অনেক কিছু উত্তরাধিকার সূত্রে পায় (যেমন চেহারা বা সম্পদ), প্রোগ্রামিংয়েও একটি ক্লাস অন্য একটি ক্লাসের বৈশিষ্ট্য বা কোড ধার নিতে পারে। যে ক্লাস থেকে বৈশিষ্ট্য নেয়া হয় তাকে বলে Parent Class, আর যে ক্লাসটি বৈশিষ্ট্য গ্রহণ করে তাকে বলে Child Class।"

### Why do we use it? (কেন এটা ব্যবহার করব?)

1.  Code Reusability: আপনাকে একই কোড বারবার লিখতে হবে না। বাবার কোড ছেলে অটোমেটিক পেয়ে যাবে।
2.  Organization: কোড অনেক বেশি গোছানো থাকে। সাধারণ বৈশিষ্ট্যগুলো এক জায়গায় থাকে, আর বিশেষ বৈশিষ্ট্যগুলো আলাদা থাকে।
3.  Time Saving: বড় সফটওয়্যার তৈরির সময় এটি প্রচুর সময় বাঁচায়।

### How does it work? (কিভাবে কাজ করে?)

-   Inheritance বুঝতে হলে ৩টি জিনিস জানতে হবে:
-   Parent Class: মেইন ক্লাস (যেমন: Vehicle)
-   Child Class: যে মেইন ক্লাসকে অনুসরণ করে (যেমন: Car বা Bike)
-   super() Function: এটি চাইল্ড ক্লাস থেকে প্যারেন্ট ক্লাসের __init__ বা অন্যান্য মেথডকে কল করতে সাহায্য করে।

#### Simple Example Code:

Inheritance মানে হলো চাইল্ড ক্লাস প্যারেন্ট ক্লাসের সব বৈশিষ্ট্য উত্তরাধিকারসূত্রে পায়। যদি চাইল্ড ক্লাসে কোনো কিছু না থাকে, পাইথন স্বয়ংক্রিয়ভাবে প্যারেন্ট ক্লাসে গিয়ে সেটা খুঁজে বের করে।

```python
class Animal:
    def eat(self):
        print("This animal is eating...")

class Dog(Animal):
    def bark(self):
        print("Dog barks!")

my_dog = Dog()
my_dog.bark() # চাইল্ড ক্লাসের নিজস্ব মেথড
my_dog.eat()  # <-- ম্যাজিক! Dog ক্লাসে 'eat' নেই, কিন্তু সে Animal থেকে এটি পেয়েছে।
```

### Method Overriding (নিজের মতো করে সাজানো)

-   যখন আপনি Dog ক্লাসে speak লিখছেন, তখন আপনি পাইথনকে বলছেন: "বাবার (Animal) কথা বলার স্টাইল আমার পছন্দ না, আমি আমার নিজের স্টাইলে কথা বলব।"
-   তাই যখন আপনি my_dog.speak() কল করেন, পাইথন প্রথমে চাইল্ড ক্লাসে খোঁজে। যদি সেখানে থাকে, তবে সেটাই চালায়। যদি না থাকে, তখন প্যারেন্ট ক্লাসে খোঁজে এবং সেটাই চালায়।

```python
# প্যারেন্ট ক্লাস (বাবা)
class Animal:
    def speak(self):
        print("Animal makes a sound")

# চাইল্ড ক্লাস (সন্তান) - ব্র্যাকেটে প্যারেন্ট ক্লাসের নাম দিতে হয়
class Dog(Animal):
    def speak(self):
        print("Dog barks! 🐶")

my_dog = Dog()
my_dog.speak() # এটি 'Dog barks' প্রিন্ট করবে
```

### Parent Class কেন দরকার?

আপনার মনে হতে পারে, আমি তো সব কোড Dog ক্লাসেই লিখতে পারতাম, তবে Animal ক্লাসের দরকার কী?

উত্তর হলো Consistency (একই নিয়ম বজায় রাখা):

ধরুন আপনার প্রোগ্রামে ১০০ ধরণের প্রাণী আছে (Dog, Cat, Cow, Tiger...)।

-   বিনা ইনহেরিটেন্স: আপনাকে ১০০ বার eat(), sleep(), breath() ফাংশনগুলো আলাদা করে লিখতে হবে।
-   ইনহেরিটেন্স সহ: আপনি শুধু একবার Animal ক্লাসে eat() লিখবেন। বাকি ১০০টি ক্লাস শুধু Animal কে ইনহেরিট করবে এবং অটোমেটিক সেই কোড পেয়ে যাবে।

```python
class Dog:
    def eat(self):
        print("Eating food...")
    def sleep(self):
        print("Sleeping...")
    def bark(self):
        print("Dog barks!")

class Cat:
    def eat(self): # একই কোড আবার লিখতে হচ্ছে
        print("Eating food...")
    def sleep(self): # একই কোড আবার লিখতে হচ্ছে
        print("Sleeping...")
    def meow(self):
        print("Cat meows!")

# এভাবে ১০০টি প্রাণীর জন্য আলাদা আলাদা eat() এবং sleep() লিখতে হবে।
```

### ইনহেরিটেন্স সহ (The Smart Way)
এখানে আমরা কমন কাজগুলো (যেমন: খাওয়া এবং ঘুমানো) একবার Animal ক্লাসে লিখব। বাকি সবাই সেটা অটোমেটিক পেয়ে যাবে।

```python
# প্যারেন্ট ক্লাস (বাবার ক্লাস)
class Animal:
    def eat(self):
        print("This animal is eating...")
    
    def sleep(self):
        print("This animal is sleeping...")

# চাইল্ড ক্লাসসমূহ (সন্তান)
class Dog(Animal): # শুধু Animal-কে ইনহেরিট করলাম
    def bark(self):
        print("Dog barks! 🐶")

class Cat(Animal): # শুধু Animal-কে ইনহেরিট করলাম
    def meow(self):
        print("Cat meows! 🐱")

class Tiger(Animal): 
    def roar(self):
        print("Tiger roars! 🐯")

# ব্যবহার করে দেখা
my_dog = Dog()
my_dog.eat()   # Animal ক্লাস থেকে এসেছে
my_dog.bark()  # Dog ক্লাসের নিজস্ব

my_tiger = Tiger()
my_tiger.eat()   # Animal ক্লাস থেকে এসেছে
my_tiger.roar()  # Tiger ক্লাসের নিজস্ব
```

1.  Dry Principle (Don't Repeat Yourself): আমরা eat() এবং sleep() মাত্র একবার লিখেছি, কিন্তু ব্যবহার করছি ১০০ বার।
2.  Easy to Update: ধরুন আপনি এখন চান সব প্রাণী খাওয়ার পর "আলহামদুলিল্লাহ" বা "Thank you" বলবে। ইনহেরিটেন্স ছাড়া আপনাকে ১০০টি ক্লাসে গিয়ে পরিবর্তন করতে হতো। কিন্তু এখন আপনি শুধু প্যারেন্ট ক্লাসে (Animal) একবার পরিবর্তন করলেই বাকি ১০০টি ক্লাসে তা অটোমেটিক আপডেট হয়ে যাবে!
3.  Focus on Specialty: Dog ক্লাসের ভেতরে আপনাকে শুধু কুকুরের বিশেষ কাজ (যেমন bark) নিয়ে চিন্তা করতে হবে, সাধারণ কাজ (যেমন eat) নিয়ে নয়।

### super() কীওয়ার্ডের ব্যবহার
মাঝে মাঝে সন্তান চায় বাবার কাজটাও করতে, আবার নিজের কিছু যোগ করতে। তখন আমরা super() ব্যবহার করি।

```python
class Parent:
    def works(self):
        print("Parent Workds hard to provide for the family.")

class Child(Parent):
    def works(self):
        super().works() # এটি বাবার 'works' মেথডকে কল করবে
        print("Specifically, the child works! 🧒")

my_child = Child()
my_child.works()
```

### Mini Project — "The Smart Vehicle Showroom"
আমরা একটি সাধারণ Vehicle ক্লাস বানাব এবং তা থেকে Car এবং ElectricCar তৈরি করব।

```python
# --- Video 13 Project: Vehicle Showroom ---

# প্যারেন্ট ক্লাস
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def display_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}")

# চাইল্ড ক্লাস (Inheritance)
class Car(Vehicle):
    def __init__(self, brand, model, fuel_type):
        # super() দিয়ে প্যারেন্ট ক্লাসের brand এবং model সেট করা হচ্ছে
        super().__init__(brand, model)
        self.fuel = fuel_type

    def drive(self):
        print(f"{self.brand} {self.model} is running on {self.fuel}!")

# অবজেক্ট তৈরি
my_car = Car("Toyota", "Corolla", "Octane")
my_car.display_info() # প্যারেন্ট ক্লাসের মেথড
my_car.drive()        # চাইল্ড ক্লাসের মেথড
```

ইনহেরিটেন্স মানে শুধু বাবার কোড বদলে ফেলা নয়, বরং বাবার তৈরি করা সব 'ফিচার' বিনা পরিশ্রমে নিজের ক্লাসে নিয়ে আসা। যদি আমরা Animal ক্লাসে ১০০০ লাইন কোড লিখি, তবে Dog(Animal) লিখলেই সেই ১০০০ লাইন কোড আমাদের Dog ক্লাসে চলে আসবে। এটাই হলো ইনহেরিটেন্সের আসল শক্তি!
