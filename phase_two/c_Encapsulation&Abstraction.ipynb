### Encapsulation & Abstraction (The Shield and the Dashboard) 🛡️🕹️

### What are they? (কি এগুলো?)

**Encapsulation (এনক্যাপসুলেশন - সুরক্ষা):**</br>
এনক্যাপসুলেশন মানে হলো আপনার ক্লাসের ডাটাকে একটি ক্যাপসুলের মতো আটকে রাখা যাতে বাইরের কেউ তা হুট করে পরিবর্তন করতে না পারে।</br>
**যেমন:** আপনার ব্যাংক ব্যালেন্স একটি ভেরিয়েবল হতে পারে, কিন্তু আপনি নিশ্চয়ই চাইবেন না যে কেউ সরাসরি balance = 0 লিখে আপনার টাকা গায়েব করে দিক!

**Abstraction (অ্যাবস্ট্রাকশন - সহজীকরণ):**</br>
অ্যাবস্ট্রাকশন মানে হলো অপ্রয়োজনীয় জটিলতা লুকিয়ে রেখে শুধু প্রয়োজনীয় অংশটুকু ইউজারকে দেখানো।</br>
**যেমন:** আপনি যখন গাড়ি চালান, আপনি শুধু জানেন এক্সিলারেটর চাপলে গাড়ি চলবে। ইঞ্জিনের ভেতরে পিস্টন কীভাবে নড়ছে তা আপনার জানার দরকার নেই। এটাই হলো অ্যাবস্ট্রাকশন।

### Why do we use it? (কেন এগুলো ব্যবহার করব?)

1. **Security (Encapsulation):** এটি ডাটাকে ভুলবশত পরিবর্তন হওয়া থেকে বাঁচায়।
2. **Simplicity (Abstraction):** ইউজারকে জটিল লজিক নিয়ে মাথা ঘামাতে হয় না, তারা শুধু ফাংশন কল করে কাজ সারতে পারে।
3. **Maintainability:** আপনি যখন ব্যাক-এন্ডে কোড পরিবর্তন করবেন, ইউজারের সামনে থাকা ইন্টারফেস একই থাকবে।

### How does it work? (কিভাবে কাজ করে?)
**How Encapsulation works (Private Variables):**</br>
পাইথনে ভেরিয়েবলের নামের আগে **Double Underscore (__)** দিলে সেটি **Private** হয়ে যায়। অর্থাৎ, ক্লাসের বাইরে থেকে সরাসরি তাকে দেখা বা পরিবর্তন করা যায় না।

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance # এটি প্রাইভেট ভেরিয়েবল

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. Current Balance: {self.__balance}")
```

**Without Encapsulation**

```python
class OpenAccount:
    def __init__(self, balance):
        self.balance = balance  # এটি Public, সবাই এটা দেখতে এবং বদলাতে পারে

    def show_balance(self):
        return self.balance
    
# --- সমস্যাটি দেখুন ---
my_acc = OpenAccount(1000)
print(f"আসল ব্যালেন্স: {my_acc.show_balance()}")

# কোনো নিয়ম ছাড়াই যে কেউ ব্যালেন্স বদলে দিচ্ছে!
my_acc.balance = -5000  # এটা কি হওয়া উচিত? টাকা কি মাইনাস হতে পারে? 
print(f"ভুল কোডের পর ব্যালেন্স: {my_acc.show_balance()}")  # এটা আসল ব্যালেন্সকে মাইনাস করে ফেলেছে!
```

**Wit Encapsulation**

```python
class SecureAccount:
    def __init__(self, balance):
        self.__balance = balance  # এটি Private, বাইরে থেকে দেখা যাবে না

    # টাকা জমা দেওয়ার একমাত্র উপায়
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount} টাকা জমা হয়েছে। বর্তমান ব্যালেন্স: {self.__balance}")
        else:
            print("ভুল! আপনি জিরো বা মাইনাস টাকা জমা দিতে পারবেন না।")

    # ব্যালেন্স দেখার জন্য আমরা একটি 'পদ্ধতি' দেব
    def show_balance(self):
        return self.__balance

my_acc = SecureAccount(1000)

# সঠিক উপায়ে টাকা জমা দিন
my_acc.deposit(1000)
my_acc.deposit(-200) # এখানে আমাদের 'নিয়ম' তাকে আটকে দেবে

print(f"সুরক্ষিত ব্যালেন্স: {my_acc.show_balance()}")


# হ্যাকার যদি সরাসরি বদলাতে চায়:
my_acc.__balance = -5000  # এটি কাজ করবে না! 
print(f"হ্যাকার কোডের পর ব্যালেন্স: {my_acc.show_balance()}")  # এটা আসলে একটা নতুন ভেরিয়েবল তৈরি করবে, আসল ব্যালেন্সকে না ছুঁয়ে।
# কারণ পাইথন বাইরের এই __balance কে অন্য কিছু মনে করবে, আসলের সাথে মিশাবে না。
```

**অ্যাবস্ট্রাকশন কী?**

ইউজার শুধু জানে deposit() ফাংশন কল করলে টাকা জমা হয়। কিন্তু ফাংশনের ভেতরে কীভাবে if-else চেক হচ্ছে বা ডাটা আপডেট হচ্ছে, সেটা ইউজারের জানার দরকার নেই।

সহজ কথা: এনক্যাপসুলেশন ডাটাকে সুরক্ষা দেয়, আর অ্যাবস্ট্রাকশন সিস্টেমকে সহজ করে দেখায়।
