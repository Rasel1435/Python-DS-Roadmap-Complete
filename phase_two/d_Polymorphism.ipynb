## Polymorphism (একই নামে বহু রূপ বা বহুরূপী ) 🎭

### What is Polymorphism? (কি এটা?)

আসসালামু আলাইকুম! 'Poly' মানে হলো অনেক (Many) আর 'Morph' মানে হলো রূপ (Form)। অর্থাৎ পলিমরফিজম মানে হলো 'একই জিনিসের বহু রূপ'।

প্রোগ্রামিংয়ের ভাষায়, যখন আমরা একই নামের ফাংশন বা মেথড আলাদা আলাদা ক্লাসে ব্যবহার করি কিন্তু তারা প্রত্যেকে নিজের মতো করে আলাদা কাজ করে, তাকেই পলিমরফিজম বলে।

**সহজ উদাহরণ:** একজন মানুষ। সে যখন অফিসে থাকে তখন সে 'এমপ্লয়ি', যখন বাসায় থাকে তখন সে 'বাবা/মা', আর যখন বাসে থাকে তখন সে 'যাত্রী'। মানুষটি একই, কিন্তু জায়গা ভেদে তার আচরণ বা কাজ আলাদা। এটাই পলিমরফিজম।

### Why do we use it? (কেন এটা ব্যবহার করব?)
1.  Uniformity: আপনাকে অনেকগুলো আলাদা ফাংশনের নাম মনে রাখতে হবে না। একটি নাম দিয়েই সব কাজ চালানো যাবে।
2. Flexibility: আপনি চাইলে সহজেই নতুন কোনো অবজেক্ট যোগ করতে পারেন, কোডের মেইন লজিক পরিবর্তন না করেই।
3. Clean Code: এটি কোডকে অনেক বেশি রিডবল এবং প্রফেশনাল করে তোলে।

### How does it work? (কিভাবে কাজ করে?)
Method Overriding: যখন একটি সাবক্লাস (Child Class) তার প্যারেন্ট ক্লাসের (Parent Class) মেথডকে আবার ডিফাইন করে, তখন তাকে Method Overriding বলে। এতে করে সাবক্লাসের মেথডটি প্যারেন্ট ক্লাসের মেথডকে ওভাররাইড করে দেয়।

Example:

```python
class Bird:
    def sound(self):
        print("Birds make different sounds")

class Parrot(Bird):
    def sound(self): # একই নাম কিন্তু কাজ আলাদা
        print("Parrot talks like humans! 🦜")

class Crow(Bird):
    def sound(self): # একই নাম কিন্তু কাজ আলাদা
        print("Crow says: Caw Caw! 🐦‍⬛")

# ব্যবহার দেখুন
birds = [Parrot(), Crow()]

for bird in birds:
    bird.sound() # একই ফাংশন কল করছি, কিন্তু আউটপুট আলাদা আসছে!
```

**Another Example:**

```python
class Human:
    def move(self):
        print("Walking on two legs... 🚶")

class Car:
    def move(self):
        print("Driving on four wheels... 🚗")

class Bird:
    def move(self):
        print("Flying in the sky... 🦅")

# পলিমরফিজমের ম্যাজিক
objects = [Human(), Car(), Bird()]

for obj in objects:
    obj.move() # ফাংশন নাম 'move' একই, কিন্তু একেক জন একেক ভাবে মুভ করছে!
```

চিন্তা করুন যদি আমাকে মানুষের জন্য walk(), গাড়ির জন্য drive(), আর পাখির জন্য fly()—সব আলাদা আলাদা মনে রাখতে হতো, তবে কোড কত কঠিন হতো! কিন্তু পলিমরফিজম আমাদের শিখিয়েছে যে, আমরা শুধু একটি নাম move() মনে রাখব, বাকিটা পাইথন বুঝে নেবে কে কীভাবে মুভ করবে।

**Mini Project — "The Global Payment System"**

আমরা একটি পেমেন্ট সিস্টেম বানাব যেখানে ইউজার 'বিকাশ', 'নগদ' বা 'কার্ড'—যেকোনো মাধ্যমেই পেমেন্ট করতে পারবে, কিন্তু আমরা শুধু একটি কমান্ড process_payment() ব্যবহার করব।

```python
# --- Video 15 Project: Universal Payment System ---

class Bkash:
    def process_payment(self, amount):
        print(f"Bkash: Processing {amount} TK. Sending OTP to your phone... ✅")

class Nagad:
    def process_payment(self, amount):
        print(f"Nagad: Processing {amount} TK. Please enter your secret PIN... ✅")

class CreditCard:
    def process_payment(self, amount):
        print(f"Card: Processing {amount} TK. Verifying with Bank Server... ✅")

# পলিমরফিজমের আসল ম্যাজিক এখানে
def make_payment(payment_obj, amount):
    payment_obj.process_payment(amount) # সে জানে না এটা কোন গেটওয়ে, সে শুধু জানে একে প্রসেস করতে হবে।

# অবজেক্ট তৈরি
user_bkash = Bkash()
user_card = CreditCard()

# একই ফাংশন ব্যবহার করে আলাদা আলাদা পেমেন্ট
make_payment(user_bkash, 5000)
make_payment(user_card, 12000)
```

### ফাংশন (Function) কী?
ফাংশন হলো কোডের একটি সেট বা ব্লক যা একটি নির্দিষ্ট কাজ করার জন্য লেখা হয়। এটি কোনো ক্লাসের ওপর নির্ভর করে না, প্রোগ্রামের যেকোনো জায়গা থেকে একে সরাসরি কল করা যায়।

**উদাহরণ:**

```python
# এটি একটি সাধারণ ফাংশন
def add_numbers(a, b):
    return a + b

# সরাসরি কল করা যাচ্ছে
result = add_numbers(10, 20)
print(f"যোগফল: {result}")
```

### মেথড (Method) কী?
মেথডও কোডের একটি সেট, কিন্তু এটি ক্লাসের ভেতরে সংজ্ঞায়িত থাকে। মেথড মূলত ওই ক্লাসের অবজেক্ট কী কী কাজ করতে পারবে তা নির্ধারণ করে। মেথড কল করতে হলে অবশ্যই একটি অবজেক্ট লাগে এবং এর ভেতর প্রথম প্যারামিটার হিসেবে self থাকতে হয়।

**উদাহরণ:**

```python
class Calculator:
    # এটি একটি মেথড (কারণ এটি ক্লাসের ভেতরে আছে)
    def multiply(self, a, b):
        return a * b

# মেথড সরাসরি কল করা যায় না, আগে অবজেক্ট তৈরি করতে হয়
my_calc = Calculator()

# অবজেক্ট দিয়ে কল করতে হচ্ছে
result = my_calc.multiply(5, 4)
print(f"গুণফল: {result}")
```

সব মেথডই আসলে ফাংশন, কিন্তু সব ফাংশন মেথড নয়।" কারণ মেথড হলো বিশেষ ধরণের ফাংশন যা ক্লাসের ভেতর বন্দী থাকে।
