### What is a Module? (কি এটা?)

আসসালামু আলাইকুম Everyone! গত ৯টি ভিডিওতে আমরা নিজেরা কোড লিখতে শিখেছি। কিন্তু দুনিয়ার সব কোড কি আমাদের নিজেদের লিখতে হবে? একদম না! Module হলো একটি ফাইল যাতে আগে থেকেই অনেক ফাংশন এবং কোড লিখে রাখা হয়েছে। যেমন: অংকের জন্য math মডিউল বা র‍্যান্ডম নাম্বারের জন্য random মডিউল। আর অনেকগুলো মডিউল যখন একসাথে থাকে, তখন তাকে বলে Package।

### Why do we use it? (কেন এটা ব্যবহার করব?)

-   Save Time: জটিল কাজ (যেমন ওদার ডাটা আনা বা ক্যালকুলেশন) আগে থেকেই কেউ করে রেখেছে, আপনি শুধু ব্যবহার করবেন।
-   Standardization: বিশ্বজুড়ে প্রোগ্রামাররা একই লাইব্রেরি ব্যবহার করে, তাই আপনার কোড সবাই বুঝবে।
-   Power: পাইথনের বিশাল ভাণ্ডার (যেমন ডাটা সায়েন্সের জন্য NumPy, Pandas) ব্যবহার করার সুযোগ।

### Mini Project — "The Secret Number & Math Power"
আমরা একটি প্রোগ্রাম বানাব যা random মডিউল ব্যবহার করে একটি সিক্রেট নাম্বার জেনারেট করবে এবং math ব্যবহার করে কিছু ক্যালকুলেশন করবে।

```python
# --- Project: Secret Number Tool ---
import random
import math

def start_game():
    secret_number = random.randint(1, 100)
    print(f"I have a secret number! Hint: The square root of it is {math.sqrt(secret_number):.2f}")
    
    guess = int(input("Guess the secret number: "))
    
    if guess == secret_number:
        print("Congratulations! You found the secret superpower!")
    else:
        print(f"Wrong! The secret was {secret_number}. Better luck next time!")

start_game()
```
