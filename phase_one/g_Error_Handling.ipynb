## Error Handling (Building Unbreakable Code) 🛡️

### What is Error Handling? (কি এটা?)
আসসালামু আলাইকুম! আমরা অনেক সুন্দর কোড লিখছি, কিন্তু ধরুন আপনার প্রোগ্রামটি চলার সময় হঠাৎ বন্ধ হয়ে গেল বা 'Crash' করল। কেন? হয়তো ইউজার ভুল ইনপুট দিয়েছে বা আপনি এমন কিছু দিয়ে ভাগ করেছেন যা সম্ভব না। এই যে অনাকাঙ্ক্ষিত ভুল বা বাগগুলো সামাল দেওয়ার পদ্ধতিই হলো Error Handling।

### Why do we use it? (কেন এটা ব্যবহার করব?)

-   Prevent Crashing: প্রোগ্রাম যেন মাঝপথে হুট করে বন্ধ না হয়ে যায়।
-   User Experience: ইউজারকে লাল রঙের ভয়ংকর এরর না দেখিয়ে একটি সুন্দর মেসেজ দেওয়া (যেমন: "দয়া করে সঠিক নাম্বার দিন")।
-   Debugging: কোথায় ভুল হচ্ছে তা সহজে খুঁজে বের করা।

### How does it work? (কিভাবে কাজ করে?)
পাইথনে এটি কাজ করে Try-Except ব্লকের মাধ্যমে। একে একটি 'Safety Net' বা নিরাপত্তা জালের মতো চিন্তা করুন।

-   Try: এখানে আমরা সেই কোডটি রাখি যা ভুল হওয়ার সম্ভাবনা আছে।
-   Except: যদি 'Try' ব্লকে কোনো ভুল হয়, তবে প্রোগ্রামটি বন্ধ না হয়ে এই ব্লকে চলে আসবে এবং আপনি যা বলবেন তাই করবে।

```python
# একটি সহজ ভাগের উদাহরণ
try:
    number = int(input("Enter a number to divide 100: "))
    result = 100 / number
    print(f"Result is: {result}")
except ZeroDivisionError:
    print("Error: You cannot divide by zero!")
except ValueError:
    print("Error: Please enter a valid number, not text!")
```

### The "Finally" Block
Finally ব্লকটি হলো পাইথন এরর হ্যান্ডলিংয়ের এমন একটি অংশ যা অবশ্যই রান করবে, আপনার কোডে এরর আসুক আর না আসুক।
সহজ ভাষায় বলতে গেলে: এটি হলো আপনার কোডের "শেষ কথা" বা "Cleanup Crew"।

### কেন আমরা এটি ব্যবহার করি? (Why we use it?)
বাস্তব জীবনে প্রোগ্রামিং করার সময় আমাদের কিছু কাজ সবসময় শেষ করতে হয়। যেমন:
-   ফাইল বন্ধ করা: আপনি একটি ফাইল ওপেন করে ডাটা পড়ছিলেন, মাঝপথে এরর আসলো। ফাইলটি কিন্তু ওপেনই থেকে যাবে যদি না আপনি তা বন্ধ করেন।
-   ডাটাবেজ কানেকশন ডিসকানেক্ট করা: কাজ শেষ হোক বা না হোক, সার্ভারের সাথে কানেকশন কেটে দেওয়া জরুরি।
-   রিসোর্স ফ্রি করা: মেমোরি খালি করা।

### সহজ উদাহরণ (Easy Example)
ধরুন, আপনি ল্যাপটপে একটি গেম খেলছেন:

Try: গেমটি খেলা শুরু করলেন।

Except: গেমটি ক্র্যাশ করল বা লোড হলো না।

Finally: যাই হোক না কেন, ল্যাপটপটি বন্ধ (Shut down) করা জরুরি।

```python
try:
    print("Opening a data file...")
    # কিছু কোড...
except Exception as e:
    print(f"Something went wrong: {e}")
finally:
    print("Closing the file anyway. Done!")
```

### Mini Project — "Safe Calculator for Everyone"
চলুন একটি ক্যালকুলেটর বানাই যা ভুল ইনপুট দিলেও ক্র্যাশ করবে না।

```python
# --- Project: Smart & Safe Calculator ---

def safe_calculator():
    print("--- Welcome to Safe Calc Pro ---")
    
    try:
        # ইউজার থেকে ইনপুট নেয়া
        num1 = float(input("Enter first number: "))
        operation = input("Enter operation (+, -, *, /): ")
        num2 = float(input("Enter second number: "))

        # লজিক এবং ক্যালকুলেশন
        if operation == "+":
            result = num1 + num2
        elif operation == "-":
            result = num1 - num2
        elif operation == "*":
            result = num1 * num2
        elif operation == "/":
            result = num1 / num2
        else:
            result = "Invalid operation!"
        
        print(f"\nThe Result is: {num1} {operation} {num2} = {result}")

    except ValueError:
        print("\nError: দয়া করে শুধু সংখ্যা (Numbers) ইনপুট দিন।")
    except ZeroDivisionError:
        print("\nError: শূন্য (0) দিয়ে ভাগ করা সম্ভব নয়।")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")

    finally:
        # এই অংশটি সবসময় রান করবেই
        print("\n--- ক্যালকুলেশন সেশন শেষ হলো। ধন্যবাদ! ---")

safe_calculator()
```

২. Except: "যদি ইউজার ভুল করে কোনো অক্ষর (Text) দেয় বা শূন্য দিয়ে ভাগ করতে চায়, তবে আমাদের প্রোগ্রাম ক্র্যাশ করবে না; বরং একটি সুন্দর ওয়ার্নিং মেসেজ দিবে।"

৩. Finally: "সবশেষে খেয়াল করুন, ইউজার সঠিক ইনপুট দিক বা ভুল করুক—আমাদের finally ব্লকের মেসেজটি কিন্তু সবসময়ই দেখা যাচ্ছে। এর মানে হলো সেশনটি সাকসেসফুলি ক্লোজ হয়েছে।"
