## Loops & Iteration (Automate Everything) 🔄

#### What is a Loop? 🤔
সহজ ভাষায় বলতে গেলে, Loop (লুপ) হলো প্রোগ্রামিংয়ের এমন একটি পদ্ধতি যা দিয়ে একই কাজ বারবার (Repeat) করানো যায়।

ধরুন, আপনাকে বলা হলো আপনার নাম ১০০ বার প্রিন্ট করতে। আপনি কি ১০০ বার print("Your Name") লিখবেন? অবশ্যই না! আপনি একটি Loop ব্যবহার করবেন যা মাত্র ৩ লাইন কোডেই আপনার নাম ১০০ বার প্রিন্ট করে দিবে।

#### লুপ কেন ব্যবহার করবেন?

-   সময় বাঁচাতে: একই কোড বারবার লিখতে হয় না।
-   অটোমেশন: হাজার হাজার ডাটা (যেমন ১০০০ স্টুডেন্টের রেজাল্ট) এক নিমিষেই প্রসেস করা যায়।
-   কোড ক্লিন রাখা: কম কোডে বড় কাজ করা যায়।

#### পাইথনে লুপ প্রধানত ২ প্রকার:
-   ১. For Loop (যখন আমরা জানি কতবার ঘুরতে হবে)
যদি আপনি জানেন যে কাজটা ১০ বার বা ২০ বার করতে হবে, তখন For Loop ব্যবহার করা হয়। এটি সাধারণত কোনো লিস্ট বা রেঞ্জ (Range) এর ওপর কাজ করে।

-   ২. While Loop (যতক্ষণ শর্ত সত্যি থাকবে)
এটি ততক্ষণ কাজ করতে থাকে যতক্ষণ একটি নির্দিষ্ট কন্ডিশন বা শর্ত True থাকে। শর্তটি False হলেই লুপটি থেমে যায়।

#### বাস্তব জীবনের উদাহরণ (Real-world Example):

-   ফ্যান চলা: যতক্ষণ সুইট অন (True), ততক্ষণ ফ্যানটি ঘুরতে থাকে (Loop)। সুইচ অফ (False) করলে লুপ বন্ধ হয়ে যায়।

-   গান শোনা: মিউজিক প্লেয়ারে যখন আপনি 'Repeat' অন করেন, তখন গানটি লুপে চলতে থাকে।

লুপ হলো প্রোগ্রামিংয়ের সেই জাদুকরী শক্তি যা আমাদের ঘণ্টার কাজ সেকেন্ডে করে দেয়। ডাটা সায়েন্সে যখন আমরা লাখ লাখ ডাটা নিয়ে কাজ করব, তখন এই লুপ-ই হবে আমাদের সেরা বন্ধু

### The For Loop (Definite Iteration)
যখন আমরা আগে থেকেই জানি যে একটি কাজ কতবার করতে হবে (যেমন একটি লিস্টের সব আইটেম প্রিন্ট করা), তখন আমরা ব্যবহার করি For Loop।

```python
# লিস্টের ওপর লুপ চালানো
fruits = ["Apple", "Banana", "Mango"]
for fruit in fruits:
    print(f"I want to eat {fruit}")

# range() ফাংশন ব্যবহার করে লুপ (১ থেকে ৫ পর্যন্ত)
for i in range(1, 6):
    print(f"Counting: {i}")
```

### The While Loop (Indefinite Iteration)
যখন আমরা জানি না কাজটা কতবার হবে, বরং একটি শর্ত সত্যি থাকা পর্যন্ত কাজটা চলতেই থাকবে—তখন আমরা ব্যবহার করি While Loop।

```python
# যতক্ষণ ব্যাটারি ৫% এর উপরে, ততক্ষণ ফোন চলবে
battery = 10

while battery > 5:
    print(f"Phone is running... Battery: {battery}%")
    battery -= 1  # ব্যাটারি এক এক করে কমছে

print("Battery low! Please charge.")
```

### Break vs. Continue (Loop Control)
* Break: লুপের মাঝপথে যদি আমাদের মনে হয় আর সামনে যাওয়ার দরকার নেই, তখন আমরা লুপ থেকে বের হয়ে আসি (Stop)।"

*   Continue: যদি বর্তমান আইটেমটি বাদ দিয়ে সরাসরি পরের আইটেমে যেতে চাই (Skip)।

```python
# নম্বর পেলে লুপ বন্ধ হবে
for n in range(1, 10):
    if n == 5:
        break 
    print(n)

# নম্বরটি স্কিপ হবে
for x in range(1, 5):
    if x == 3:
        continue
    print(f"Value: {x}")
```

### Mini Project — "Smart Sensor Data Cleaner"
In Data Science, we often get "bad data." Let's build a loop that cleans a list of sensor readings!

```python
# --- Project: Smart Sensor Data Cleaner ---

# সেন্সর থেকে আসা ডাটা (কিছু ডাটা ভুল বা None)
raw_readings = [22.5, 23.0, -999, 24.5, 22.8, -999, 25.1]
clean_data = []

print("Starting Data Cleaning...")

for data in raw_readings:
    # -999 মানে সেন্সরে সমস্যা ছিল, এটা স্কিপ করতে হবে
    if data == -999:
        print("Skipping faulty data...")
        continue
    
    clean_data.append(data)

print(f"Cleaned Readings: {clean_data}")
print(f"Average Temp: {sum(clean_data)/len(clean_data):.2f}°C")
```
