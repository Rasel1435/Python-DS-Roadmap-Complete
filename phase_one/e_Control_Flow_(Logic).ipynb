## Control Flow (The Brain of Your Code) 🧠
আসসালামু আলাইকুম! গত ৫টি ভিডিওতে আমরা ডাটা স্টোর করা এবং সাজানো শিখেছি। কিন্তু আমাদের প্রোগ্রামগুলো এখনো নিজে থেকে কোনো সিদ্ধান্ত নিতে পারে না। আজ আমরা শিখব Control Flow। সহজ কথায়, এটি আপনার কোডের 'মস্তিষ্ক'। আমরা শিখব কীভাবে if, elif, এবং else ব্যবহার করে প্রোগ্রামকে সিদ্ধান্ত নিতে হয়—ঠিক যেমন আমরা বাস্তব জীবনে নিয়ে থাকি।

### The logic Breakdown
-   if: প্রথম শর্ত। (যদি বৃষ্টি হয়...)
-   elif (else if): যদি প্রথমটা না হয়ে অন্য কিছু হয়। (...নাকি যদি রোদেলা আবহাওয়া থাকে?)
-   else: যদি কোনো শর্তই না মিলে। (...তবে যা ইচ্ছে করো!)

```python
# একটি সহজ উদাহরন: ইন্টারভিউ সিলেকশন
score = 69

if score >= 90:
    print("Excellent! You are selected for the Lead role.")
elif score >= 70:
    print("Good job! You are selected for the Junior role.")
elif score >= 50:
    print("You are selected for the Internship role.")
else:
    print("Sorry, keep practicing for the next time.")
```

### Nested If (শর্তের ভেতর শর্ত)
কখনো কখনো আমাদের একটি শর্ত সত্যি হওয়ার পর আরও একটি শর্ত চেক করতে হয়। একে বলে Nested If।

```python
is_member = True
purchase_amount = 600

if is_member:
    if purchase_amount > 500:
        print("You get a 20% discount!")
    else:
        print("You get a 10% discount!")
else:
    print("No discount for non-members.")
```

### Mini Project — "Smart AI Weather Assistant"

```python
# --- Project: Smart AI Weather Assistant ---

print("--- Welcome to AI Weather Assistant ---")

temperature = float(input("Enter today's temperature in Celsius: "))
print(f"Current temperature: {temperature}°C")

if temperature > 35:
    print("Suggestion: It's too hot! Stay hydrated and stay indoors. ☀️")
elif temperature >= 25 and temperature <= 35:
    print("Suggestion: The weather is pleasant. Great for a walk! 🌤️")
elif temperature >= 15 and temperature < 25:
    print("Suggestion: It's a bit chilly. Wear a light jacket. 🍃")
else:
    print("Suggestion: It's freezing! Stay warm and drink coffee. ❄️")

# Logic check for rainy weather
is_raining = input("Is it raining? (yes/no): ").lower()
if is_raining == "yes":
    print("It's raining! Don't forget your umbrella! ☔")
```
