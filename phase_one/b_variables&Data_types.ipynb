## Variables & Data Types (Deep Dive)

### What is a Variable? (The Concept)

"ভ্যারিয়েবলকে চিন্তা করুন একটি বক্সের মতো, যার ভেতর আপনি কোনো তথ্য বা ডাটা জমা রাখতে পারেন। এবং পরে ওই বক্সের নাম ধরে ডাক দিলে আপনি তথ্যটি পেয়ে যাবেন।"

```python
# Example
name = "Python"    # এখানে 'name' হলো বক্সের নাম, আর "Python" হলো ডাটা
version = 2026
```

```python
print(name)
print(version)
```

### Basic Data Types (The 4 Pillars)

"পাইথনে অনেক ধরণের ডাটা আছে, তবে ডাটা সায়েন্সের শুরুতে আমাদের এই ৪টি প্রধান টাইপ বুঝতে হবে:"

1. Integer (পূর্ণসংখ্যা)</br>
    উদাহরণ: বয়স, রোল নম্বর, বা পপুলেশন।</br>
    age = 25

2. Float (দশমিক সংখ্যা)</br>
    দাহরণ: প্রাইস, জিপিএ, বা তাপমাত্রা।</br>
    price = 99.99

3. String (লেখা বা টেক্সট)</br>
    উদাহরণ: নাম, ইমেইল, বা মেসেজ। (অবশ্যই কোটেশন "" এর ভেতর হতে হবে)</br>
    message = "Hello Learners"

4. Boolean (হ্যাঁ/না)</br>
    উদাহরণ: ইউজার কি লগইন করা? ডাটা কি ক্লিন?</br>
    is_valid = True</br>
    is_error = False


### Checking Data Types (type() function)

"যখন আপনি কোনো ভ্যারিয়েবল তৈরি করেন, তখন পাইথন স্বয়ংক্রিয়ভাবে তার ডাটা টাইপ নির্ধারণ করে দেয়। তবে আপনি চাইলে type() ফাংশন ব্যবহার করে সেটা চেক করতে পারেন।পাইথনে কোনো ভ্যারিয়েবল কোন টাইপের ডাটা হোল্ড করছে, তা দেখার জন্য আমরা type() ফাংশন ব্যবহার করা হয়।"

```python
# Example:
x = 10
y = 10.5
z = "Data Science"
```

```python
print(type(x)) # Output: <class 'int'>
print(type(y)) # Output: <class 'float'>
print(type(z)) # Output: <class 'str'>
```

### Dynamic Typing & Multiple Assignment (Pro Tips)

"পাইথনের একটি স্পেশালিটি হলো এটি Dynamically Typed ল্যাঙ্গুয়েজ। অর্থাৎ আপনাকে আগে থেকে বলে দিতে হয় না এটা কি টাইপ।"

```python
# Example:
# এক লাইনে একাধিক ভ্যারিয়েবল
user_name, user_id, is_active = "Arif", 101, True
print(user_name, user_id, is_active)
```
