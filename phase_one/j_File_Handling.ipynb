## File Handling (Saving & Loading Data) 📂

### What is File Handling? (কি এটা?)
আসসালামু আলাইকুম Everyone! গত ১০টি ভিডিওতে আমরা যা ইনপুট দিয়েছি, প্রোগ্রাম বন্ধ করার সাথে সাথে তা হারিয়ে গেছে। কিন্তু আমরা যদি চাই আমাদের ডাটা আজীবনের জন্য সেভ করে রাখতে? এখানেই আসে File Handling। এটি এমন একটি প্রসেস যার মাধ্যমে পাইথন আপনার কম্পিউটার থেকে ফাইল পড়তে (Read) পারে এবং নতুন ফাইলে ডাটা লিখতে (Write) পারে।

### Why do we use it? (কেন এটা ব্যবহার করব?)

-   Persistence: প্রোগ্রাম বন্ধ করলেও ডাটা হারাবে না।
-   Data Science Foundation: রিয়েল ওয়ার্ল্ডে ডাটা সেভ থাকে .csv, .json বা .txt ফাইলে। এগুলো হ্যান্ডেল করা শিখতেই হবে।
-   Automation: হাজার হাজার ফাইলের নাম পরিবর্তন করা বা ডাটা ট্রান্সফার করা সহজ হয়।

### How does it work? (কিভাবে কাজ করে?)
ফাইল হ্যান্ডলিং মূলত ৩টি ধাপে কাজ করে: Open -> Process (Read/Write) -> Close.

####  ফাইলে লিখা (Writing)

```python
# Code Examples:
with open("../data/note.txt", "w") as file:
    file.write("Python is amazing!\nI am learning File Handling.")
```

####  ফাইল থেকে পড়া (Reading)

```python
# Code Examples:
with open("../data/note.txt", "r") as file:
    content = file.read()
    print(content)
```

### Types of Files (Text, CSV, JSON)
1. Text (.txt): সাধারণ লেখার জন্য।
2. CSV (.csv): ডাটা সায়েন্সের জন্য সবচেয়ে জরুরি (এক্সেল ফাইলের মতো)।
3. JSON (.json): ইন্টারনেটে ডাটা আদান-প্রদান বা ডিকশনারি সেভ করার জন্য।

#### Example of CSV:

```python
import csv

# CSV ফাইলে ডাটা সেভ করা
with open("../data/students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Score"])
    writer.writerow(["Arif", 95])
```

### Mini Project — "My Daily Journal App"
আমরা একটি ছোট অ্যাপ বানাব যা ইউজারের প্রতিদিনের নোট একটি ফাইলে সেভ করে রাখবে।

The Project Code:

```python
# --- Project: Simple Journal ---
import datetime

def write_journal():
    note = input("Write your thoughts for today: ")
    date = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # 'a' মানে Append - এটি আগের লেখা না মুছে নিচে নতুন লেখা যোগ করবে
    with open("../data/my_journal.txt", "a") as file:
        file.write(f"[{date}] - {note}\n")
    
    print("Journal saved successfully!")

def read_journal():
    print("\n--- Your Past Journals ---")
    try:
        with open("../data/my_journal.txt", "r") as file:
            print(file.read())
    except FileNotFoundError:
        print("No journal entries found yet.")

# রান করা
write_journal()
read_journal()
```
