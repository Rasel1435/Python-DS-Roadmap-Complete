### Arithmetic Operators (à¦—à¦¾à¦£à¦¿à¦¤à¦¿à¦• à¦•à¦¾à¦œ)
"à¦à¦—à§à¦²à§‹ à¦†à¦®à¦°à¦¾ à¦›à§‹à¦Ÿà¦¬à§‡à¦²à¦¾à¦° à¦…à¦‚à¦•à§‡à¦° à¦®à¦¤à§‹à¦‡ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¿à¥¤"

```python
# Arithmetic Operators
a = 15
b = 4

print(a + b)  # à¦¯à§‹à¦— (19)
print(a - b)  # à¦¬à¦¿à§Ÿà§‹à¦— (11)
print(a * b)  # à¦—à§à¦£ (60)
print(a / b)  # à¦­à¦¾à¦— (3.75)
print(a // b) # à¦«à§à¦²à§‹à¦° à¦¡à¦¿à¦­à¦¿à¦¶à¦¨ - à¦¦à¦¶à¦®à¦¿à¦• à¦›à¦¾à§œà¦¾ à¦­à¦¾à¦—à¦«à¦² (3)
print(a % b)  # à¦®à¦¡à§à¦²à¦¾à¦¸ - à¦­à¦¾à¦—à¦¶à§‡à¦· (3)
print(a ** b) # à¦ªà¦¾à¦“à§Ÿà¦¾à¦° à¦¬à¦¾ à¦à¦•à§à¦¸à¦ªà§‹à¦¨à§‡à¦¨à§à¦Ÿ (15^4)
```

### Project : Smart Budget Calculator
à¦†à¦®à¦°à¦¾ à¦¤à§‹ à¦…à¦ªà¦¾à¦°à§‡à¦Ÿà¦° à¦¶à¦¿à¦–à¦²à¦¾à¦®, à¦•à¦¿à¦¨à§à¦¤à§ à¦¬à¦¾à¦¸à§à¦¤à¦¬à§‡ à¦à¦—à§à¦²à§‹ à¦•à§‹à¦¥à¦¾à§Ÿ à¦•à¦¾à¦œà§‡ à¦²à¦¾à¦—à§‡? à¦šà¦²à§à¦¨ à¦à¦•à¦Ÿà¦¾ Smart Budget Analyzer à¦¬à¦¾à¦¨à¦¾à¦‡à¥¤ à¦à¦–à¦¾à¦¨à§‡ à¦†à¦®à¦°à¦¾ à¦‡à¦‰à¦œà¦¾à¦° à¦¥à§‡à¦•à§‡ à¦‡à¦¨à¦•à¦¾à¦® à¦†à¦° à¦–à¦°à¦š à¦‡à¦¨à¦ªà§à¦Ÿ à¦¨à¦¿à¦¬à¥¤ à¦à¦°à¦ªà¦° Arithmetic Operator à¦¦à¦¿à§Ÿà§‡ à¦¸à§‡à¦­à¦¿à¦‚à¦¸ à¦•à§à¦¯à¦¾à¦²à¦•à§à¦²à§‡à¦Ÿ à¦•à¦°à¦¬ à¦à¦¬à¦‚ Comparison à¦“ Logical Operator à¦¦à¦¿à§Ÿà§‡ à¦¦à§‡à¦–à¦¬ à¦‡à¦‰à¦œà¦¾à¦°à§‡à¦° à¦¸à§‡à¦­à¦¿à¦‚à¦¸ à¦•à¦¿ à§¨à§¦% à¦à¦° à¦‰à¦ªà¦°à§‡ à¦•à¦¿ à¦¨à¦¾ à¦…à¦¥à¦¬à¦¾ à¦‡à¦‰à¦œà¦¾à¦° à¦•à¦¿ à¦‹à¦£à§‡ (Debt) à¦†à¦›à§‡ à¦•à¦¿ à¦¨à¦¾à¥¤ à¦à¦Ÿà¦¿à¦‡ à¦¹à¦²à§‹ à¦¡à¦¾à¦Ÿà¦¾ à¦…à§à¦¯à¦¾à¦¨à¦¾à¦²à¦¾à¦‡à¦¸à¦¿à¦¸à§‡à¦° à¦ªà§à¦°à¦¥à¦® à¦§à¦¾à¦ª!"

```python
# --- Mini Project: Smart Budget Analyzer ---

print("--- Welcome to Smart Budget Analyzer ---")

# Input Data
income = float(input("Enter your monthly income: "))
expense = float(input("Enter your total expenses: "))

# Arithmetic Operations (Calculating Savings)
savings = income - expense
savings_percentage = (savings / income) * 100

# Comparison & Logical Operators (Analysis)
is_saving_good = savings_percentage >= 20
is_in_debt = expense > income

# Final Report
print(f"\nYour Monthly Income: {income}")
print(f"Your Total Savings: {savings}")
print(f"Savings Percentage: {savings_percentage}%")
print(f"Is your saving habit good? (20% or more): {is_saving_good}")
print(f"Are you in debt?: {is_in_debt}")

# Logical Check for a Warning
needs_financial_plan = is_in_debt or (savings_percentage < 5)
print(f"Do you need a financial plan?: {needs_financial_plan}")
```
