## Functions (Your Personal Coding Tools) 🛠️
### What is a Function? (কি এটা?)

আসসালামু আলাইকুম! গত ৮টি ভিডিওতে আমরা যা লিখেছি, তা ছিল ওয়ান-টাইম কোড। কিন্তু ধরুন, একটি কাজ আপনার প্রোগ্রামে বারবার করা দরকার—আপনি কি বারবার একই কোড লিখবেন? অবশ্যই না! Function হলো কোডের একটি ছোট টুকরো বা 'প্যাকেট', যা একটি নির্দিষ্ট কাজ করে। আপনি একবার এটি তৈরি করবেন এবং যতবার খুশি ব্যবহার করবেন।

### Why do we use it? (কেন এটা ব্যবহার করব?)

-   Reusability: একবার লিখুন, বারবার ব্যবহার করুন (Write Once, Use Anywhere).
-   Organization: বড় কোডকে ছোট ছোট ভাগে ভাগ করে সাজিয়ে রাখা যায়।
-   Easy to Fix: যদি কোনো ভুল হয়, শুধু ফাংশনটি ঠিক করলেই সব জায়গায় ঠিক হয়ে যাবে।

### How does it work? (কিভাবে কাজ করে?)

ফাংশনকে একটি Juicer Machine-এর মতো চিন্তা করুন:
-   Arguments (Input): আপনি ফল দিচ্ছেন।
-   Logic (Processing): মেশিনটি জুস বানাচ্ছে।
-   Return (Output): আপনি গ্লাসে জুস পাচ্ছেন।

### ফাংশন ডিক্লেয়ার করা (How to Declare)
ফাংশন তৈরি করার জন্য পাইথনকে একটি সিগন্যাল দিতে হয়, আর সেই সিগন্যাল হলো def (যার মানে Definition)।

```python
def function_name(parameter):
    # এখানে কাজ (Logic) হবে
    return #result
```

-   def: পাইথনকে বলা হচ্ছে আমি একটি ফাংশন বানাচ্ছি।
-   function_name: আপনার মেশিনের নাম (যেমন: juice_maker)।
-   parameter: মেশিনে আপনি কী ইনপুট দিচ্ছেন (যেমন: orange)।

```python
# ফাংশন তৈরি (Defining)
def greet_user(name): # 'name' হলো Argument
    return f"Hello {name}, Welcome to our series!" # 'return' হলো রেজাল্ট

# ফাংশন কল করা (Calling)
message = greet_user("Rasel")
print(message)
```

```python
greet_user("Saimon")
```

### Return কী এবং এটি কীভাবে কাজ করে? (What is Return)

মনে করুন, আপনি দোকানে গিয়ে টাকা দিলেন এবং বললেন একটি চকলেট দিতে। দোকানদার চকলেটটি পকেটে ভরে রাখল কিন্তু আপনাকে দিল না—এটা হলো Print (সে শুধু দেখালো)। আর দোকানদার যদি চকলেটটি আপনার হাতে ফেরত দেয় যাতে আপনি সেটা নিয়ে বাড়িতে গিয়ে খেতে পারেন—সেটাই হলো Return।

```python
def add(a, b):
    print(a + b)

# এখানে আপনি রেজাল্টটি অন্য কোথাও ব্যবহার করতে পারবেন না।
```

```python
# Return সহ (With Return):
def add(a, b):
    return a + b

# এখন আপনি রেজাল্টটি একটি ভেরিয়েবলে জমা রাখতে পারবেন।
total = add(10, 20)
print(total * 2) # রেজাল্ট নিয়ে আরও কাজ করা যাচ্ছে!
```

```python
add(10, 200)
```

### ফাংশন কীভাবে লিখবেন এবং রি-ইউজ করবেন? (Write & Reuse)

```python
# ফাংশন লিখা (Write)
def calculate_tax(salary):
    tax = salary * 0.10 # ১০% ট্যাক্স
    return tax

# ফাংশন রি-ইউজ করা (Reuse)
# ভিন্ন ভিন্ন মানুষের জন্য একই ফাংশন বারবার কল করছি
arif_tax = calculate_tax(50000)
rahim_tax = calculate_tax(80000)
sumit_tax = calculate_tax(35000)

print(f"Arif's Tax: {arif_tax}")
print(f"Rahim's Tax: {rahim_tax}")
print(f"Sumit's Tax: {sumit_tax}")
```

```python
calculate_tax(20000)
```

### (Summary)

-   Declaration: def দিয়ে শুরু হবে, শেষে একটি কোলন : থাকবে।
-   Logic: ফাংশনের ভেতরের কোডগুলো ৪ স্পেস ডানে (Indentation) সরালে পাইথন বুঝবে এগুলো ফাংশনের অংশ।
-   Return: এটা হলো ফাংশনের আউটপুট যা আমরা পরে অন্য কাজে ব্যবহার করতে পারি।
-   Calling: ফাংশনের নাম ধরে ডাক দিলেই সে কাজ শুরু করে দিবে।
