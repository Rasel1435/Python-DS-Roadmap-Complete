 কিন্তু রিয়েল লাইফ ডাটা সায়েন্সে আমাদের এমন কিছু দরকার হয় যেখানে একটি 'Key' এর বিপরীতে একটি 'Value' থাকবে। যেমন: আপনার ফোনের কন্টাক্ট লিস্ট—নাম (Key) দিয়ে সার্চ করলে নাম্বার (Value) পাওয়া যায়। একেই বলে Dictionary। 

আর যদি আমাদের শুধু ইউনিক বা অদ্বিতীয় ডাটা দরকার হয়, তবে আছে Sets।সেট হলো এমন একটি ঝুড়ি যেখানে আপনি একই জিনিস দুইবার রাখতে পারবেন না। এটি অটোমেটিক ডুপ্লিকেট ডাটা ডিলিট করে দেয়। ডাটা ক্লিনিংয়ের সময় এটি ম্যাজিকের মতো কাজ করে।

আজ আমরা এই দুটি গুরুত্বপূর্ণ ডাটা স্ট্রাকচার শিখব।"

### Python Dictionaries (Key-Value Pairs)
"ডিকশনারিকে চিন্তা করুন একটি বাস্তব ডিকশনারির মতো। যেখানে একটি 'Word' (Key) থাকে এবং তার একটি 'Meaning' (Value) থাকে। ডাটা সায়েন্সে একে আমরা বলি Key-Value Pair।"

```python
# ডিকশনারি তৈরি করা (Curly Braces {} ব্যবহার হয়)
student = {
    "name": "Arif",
    "age": 22,
    "course": "Data Science"
}

# ডাটা এক্সেস করা (Key দিয়ে call দিতে হয়)
print(student["name"])  # Output: Arif

# নতুন ডাটা যোগ বা আপডেট করা
student["grade"] = "A+"
student["age"] = 23 # ভ্যালু আপডেট করা
student.pop("age") # "age" key-ti remove hoye jabe
print(student)
```

### Dictionary Methods (Pro Tips)
"ডিকশনারি থেকে শুধু কি (Keys) বা শুধু ভ্যালু (Values) বের করার জন্য কিছু সহজ মেথড আছে।"

```python
print(student.keys())   # সব চাবি বা Key দেখাবে
print(student.values()) # সব তথ্য বা Value দেখাবে
print(student.items())  # জোড়ায় জোড়ায় (Key, Value) দেখাবে
```

### Python Sets (Unique Items)
"সেট হলো এমন একটি কালেকশন যেখানে কোনো ডুপ্লিকেট ডাটা থাকতে পারে না। এটি আন-অর্ডার্ড, অর্থাৎ এর কোনো নির্দিষ্ট ইনডেক্স নেই। ডাটা সায়েন্সে যখন আমাদের থেকে ডুপ্লিকেট ভ্যালু বাদ দিতে হয়, তখন সেট ব্যবহার করা হয়।"

```python
# সেট তৈরি করা
my_set = {1, 2, 2, 3, 4, 4, 5}
print(my_set) # Output: {1, 2, 3, 4, 5} (ডুপ্লিকেট থাকবে না)

# সেটে ডাটা যোগ ও বিয়োগ
my_set.add(6)
my_set.remove(3)
my_set.discard(10) # যদি ডাটা না থাকে তবেও error দিবে না
my_set.update([7, 8, 9]) # একসাথে একাধিক ডাটা যোগ করা
print(my_set) # Output: {1, 2, 4, 5, 6, 7, 8, 9}

# সেট অপারেশন (Union & Intersection)
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print(set_a.union(set_b))        # সব ডাটা (১,২,৩,৪,৫)
print(set_a.intersection(set_b)) # শুধু কমন ডাটা (৩)
```

"ডিকশনারিতে (Dictionary) আমরা চাইলে Value পরিবর্তন বা আপডেট করতে পারি, কিন্তু Key নয়। আর সেটের (Set) ক্ষেত্রে আমরা কোনো আইটেম সরাসরি Update করতে পারি না (অর্থাৎ একটি আইটেমকে বদলে অন্যটি করা যায় না); আমরা শুধু নতুন আইটেম Add করতে পারি অথবা আগেরটি Remove করতে পারি। যদি কোনো আইটেম বদলাতে হয়, তবে আগেরটি রিমুভ করে নতুনটি অ্যাড করতে হয়।"

### Mini Project — Smart Student Database
"আমরা একটি স্টুডেন্ট ডাটাবেজ বানাব যেখানে ডিকশনারিতে ইনফো থাকবে এবং একটি সেটের মাধ্যমে আমরা ইউনিক কোর্সগুলো ট্র্যাক করব।"

```python
# --- Mini Project: Student Database ---
student_data = {
    "id": 101,
    "name": "Rahim",
    "skills": ["Python", "Math", "SQL"]
}

# ইউনিক স্কিল চেক করার জন্য সেট ব্যবহার
unique_skills = {"Python", "Excel", "Python", "Tableau"} 

print(f"Student: {student_data['name']}")
print(f"Total Unique Skills: {len(unique_skills)}")
print(f"Unique Skills collected: {unique_skills}")
```

"চলুন আমরা একটি ছোট অ্যাপ্লিকেশন বানাই যা একজন ইউজারের তথ্য সেভ করবে এবং তার ডুপ্লিকেট স্কিলগুলোকে অটোমেটিক ক্লিন করে দিবে।"

```python
# --- Project 2: Contact & Unique Skill Manager ---

print("--- Welcome to Smart Profile Manager ---")

# ইউজার প্রোফাইল (Dictionary)
profile = {
    "id": 1,
    "name": "Naim",
    "phone": "01644425997"
}

# স্কিল লিস্ট (এখানে ডুপ্লিকেট ইনপুট থাকতে পারে)
raw_skills = ["Python", "Excel", "SQL", "Python", "Tableau", "Excel"]

# ডুপ্লিকেট রিমুভ করা (Using Set)
clean_skills = set(raw_skills)

# প্রোফাইলে ক্লিন করা স্কিলগুলো অ্যাড করা
profile_skills = list(clean_skills) 

# আউটপুট দেখানো
print(f"\nStudent Name: {profile['name']}")
print(f"Total Unique Skills: {len(profile_skills)}")
print(f"Skills List: {profile_skills}")
```
