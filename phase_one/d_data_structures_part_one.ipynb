### Python Lists (The Flexible Bag)

"লিস্টকে চিন্তা করুন একটি বাজারের ব্যাগের মতো, যেখানে আপনি একের পর এক অনেকগুলো জিনিস রাখতে পারেন এবং চাইলে সেগুলো বদলাতেও পারেন।"

```python
# Example
# লিস্ট তৈরি করা
fruits = ["Apple", "Banana", "Mango", "Orange"]

# ইনডেক্সিং (Indexing) - মনে রাখবেন পাইথনে গুনা শুরু হয় ০ থেকে!
print(fruits[0])  # Output: Apple
print(fruits[-1]) # শেষেরটি দেখাবে: Orange

# স্লাইসিং (Slicing) - একগুচ্ছ ডাটা আলাদা করা
# [start : stop] -> স্টপ ইনডেক্স ইনক্লুড হবে না
print(fruits[1:3]) # Output: ['Banana', 'Mango']
```

### Common List Methods
"লিস্টে ডাটা যোগ করা বা মোছার জন্য পাইথনে কিছু বিল্ট-ইন মেথড আছে।"

```python
prices = [100, 200, 600, 300]

prices.append(400)      # শেষে নতুন ডাটা যোগ করা
prices.insert(1, 150)   # নির্দিষ্ট জায়গায় যোগ করা
prices.remove(200)      # ডাটা ডিলিট করা
prices.sort()           # ছোট থেকে বড় সাজানো
print(prices)
```

### Tuples (The Locked Box)
"টুপল (Tuple) হলো লিস্টের মতো একটি ডেটা কালেকশন, কিন্তু এর মূল বৈশিষ্ট্য হলো এটি Immutable। অর্থাৎ, একবার একটি টুপল তৈরি করলে আপনি তার ভেতরের আইটেম পরিবর্তন, যোগ বা ডিলিট করতে পারবেন না। একে চিন্তা করুন একটি 'সিল করা বক্সের' মতো।"

Why We Use Tuples?

-   Data Integrity (নিরাপত্তা): আপনি যদি এমন কিছু ডেটা রাখতে চান যা দুর্ঘটনাবশতও চেঞ্জ হওয়া উচিত না (যেমন: আপনার নাম, জন্ম তারিখ, বা দেশের নাম), তখন টুপল ব্যবহার করা নিরাপদ।

-   Performance (গতি): লিস্টের তুলনায় টুপল অনেক দ্রুত কাজ করে এবং মেমোরি কম খরচ করে। ডাটা সায়েন্সে যখন আমরা মিলিয়ন মিলিয়ন ডেটা নিয়ে কাজ করি, তখন টুপল ব্যবহার করা এফিসিয়েন্ট।

```python
# টুপল ডিক্লেয়ার করা (Parentheses ব্যবহার হয়)
location = (23.8103, 90.4125) # Latitude and Longitude of Dhaka

# ডিক্লেয়ার করার পর আইটেম পরিবর্তন করার চেষ্টা করলে Error আসবে
# location[0] = 24.0  <-- TypeError: 'tuple' object does not support item assignment

# একটি মাত্র আইটেম থাকলে কমা (,) দিতে হয়, নতুবা পাইথন একে String মনে করবে
single_item = ("Data",) # Correct Tuple
not_a_tuple = ("Data")  # This is just a String!
```

Tuple Unpacking

"টুপলের সবচেয়ে মজার ফিচার হলো Unpacking। আমরা এক লাইনেই টুপলের ডেটাগুলোকে আলাদা ভ্যারিয়েবলে নিয়ে আসতে পারি।"

```python
# Tuple Unpacking
user_info = ("Arif", 25, "Data Scientist")

name, age, profession = user_info

print(name)       # Output: Arif
print(age)        # Output: 25
print(profession) # Output: Data Scientist
```

### Mini Project — Daily Expense Tracker
"চলুন আমরা আমাদের লিস্টের জ্ঞান ব্যবহার করে একটি ডেইলি এক্সপেন্স ট্র্যাকার বানাই।"

```python
# Project: Expense Tracker ---
expenses = [50, 120, 30, 200, 450]

print("Total Expenses today:", sum(expenses))
print("Highest Expense:", max(expenses))
print("Lowest Expense:", min(expenses))

# নতুন খরচ যোগ করা
new_bill = float(input("Enter new bill amount: "))
expenses.append(new_bill)

print("Updated Expenses List:", expenses)
print("New Total:", sum(expenses))
```
